```python
with open("stopwords-en.txt", encoding="latin1") as file:
    stoplist = [line.rstrip().lower() for line in file]

def preprocesamiento(texto, stemming=True):
  words = []
  texto = texto.lower()
  texto = re.sub(r'[^a-zA-Z0-9_À-ÿ]', ' ', texto)
  # tokenizar
  words = nltk.word_tokenize(texto, language='spanish')
  # filtrar stopwords
  words = [word for word in words if word not in stoplist]
  # reducir palabras (stemming)
  if stemming:
      words = [stemmer.stem(word) for word in words]
  return words
```

```python
import math
import pickle

class InvertIndex:
    def __init__(self, index_file):
        self.index_file = index_file
        self.index = {}
        self.idf = {}
        self.length = {}

    def building(self, collection_text, position_text):
        total_docs = len(collection_text)
        doc_count = {}

        # build the inverted index with the collection
        for i, row in collection_text.iterrows():
          doc_id = i
          doc = row.iloc[position_text]
          keywords = preprocesamiento(doc)
        # compute the tf
          tf_per_doc = {}

          for keyword in keywords:
            if keyword in tf_per_doc:
              tf_per_doc[keyword] += 1
            else:
              tf_per_doc[keyword] = 1
          for term in tf_per_doc:
            tf_per_doc[term] = math.log10(1 + tf_per_doc[term])

        # compute the idf
          for term, tf in tf_per_doc.items():
            if term not in self.index:
                self.index[term] = []
            self.index[term].append((doc_id, tf))

            if term in doc_count:
                doc_count[term] += 1
            else:
                doc_count[term] = 1

        for term, count in doc_count.items():
            self.idf[term] = math.log(total_docs / (count))
        # compute the length (norm)
        for doc_id in range(total_docs):
            norm = 0
            for term, postings in self.index.items():
                for doc, tf in postings:
                    if doc == doc_id:
                        norm += (tf * self.idf[term]) ** 2
            self.length[doc_id] = math.sqrt(norm) #Get sum(tf*idf^2)
        # store in disk
        with open(self.index_file, 'wb') as f:
            pickle.dump((self.index, self.idf, self.length), f)

    def retrieval(self, query, k):
        self.load_index()
        N = len(self.length)
        scores = [0] * len(self.length)
        # preprocesar la query: extraer los terminos unicos
        terms = preprocesamiento(query)
        # calcular el tf-idf del query
        tf_query = {}
        for term in terms:
            if term in tf_query:
                tf_query[term] += 1
            else:
                tf_query[term] = 1
        tfidf_query = {}
        for term, tf in tf_query.items():   
            if term in self.idf:
                tfidf_query[term] = math.log10(1 + tf) * self.idf[term]

        norm_query = math.sqrt(sum(w_tq**2 for w_tq in tfidf_query.values()))

        # aplicar similitud de coseno y guardarlo en el diccionario score
        for term, w_tq in tfidf_query.items():
            if term in self.index:
                for doc, tf_td in self.index[term]:
                    w_td = tf_td * self.idf[term] #Calculo tf_idf para wt,d
                    scores[doc] += w_td * w_tq
                    
        for d in range(N):
            if self.length[d] != 0:
                scores[d] /= self.length[d]*norm_query  # Normalización del documento y la consulta
    
        # ordenar el score de forma descendente
        result = sorted(enumerate(scores), key=lambda x: x[1], reverse=True)
        # retornamos los k documentos mas relevantes (de mayor similitud al query)
        return result[:k]

    def load_index(self):
        # load index from disk
        with open(self.index_file, 'rb') as f:
            self.index, self.idf, self.length = pickle.load(f)
```

```python
import os
import pickle
import sys
class SPIMI:
    # The structure is like:
    # {TERM: [(docID, TF), ...], ...}
    # For practice we want to create n blocks whre n is equal to 2^j
    def __init__(self, index_dir="index_blocks"):
        self.index_dir = index_dir  # Directorio donde se guardarán los bloques temporales
        self.block_counter = 0      # Contador para los bloques de índice
        self.doc_count = 0          # Contador de documentos totales
        self.idf = {}
        self.length = {}
        self.disk_limit = 2000 #Limite of the memory
        self.totalLevels = 0 #Represent te quantity total of leveles 
        self.currentL = 0 #Represent in what level we are
        self.cuntermerged = 0 #Contador que nos permite observar en que nivel "ordenacion se encuentra"

        if not os.path.exists(self.index_dir):
            os.makedirs(self.index_dir)

    def spimi_invert(self, token_stream):
        dictionary = {}
        for _, row in token_stream.iterrows():
            doc_id = row['Index']
            token = row['ProcessedText']
            
            if token not in dictionary:
                dictionary[token] = {}  # Crea un diccionario para el término
            
            if doc_id not in dictionary[token]:
                dictionary[token][doc_id] = 1  # Primera aparición del término en el documento
            else:
                dictionary[token][doc_id] += 1  # Incrementa la frecuencia del término en el documento
            dictionary_size = sys.getsizeof(dictionary)
            if dictionary_size >= self.disk_limit:
                self.write_block_to_disk(dictionary)
                dictionary.clear()
                
        if dictionary:
            self.write_block_to_disk(dictionary)

        self.totalLevels = math.ceil(math.log2(self.block_counter))
        print("La cantidad de niveles seran", self.totalLevels)
        print("Cantidad de bloques generados", self.block_counter)
        
    def write_block_to_disk(self, dictionary):
        sorted_terms = dict(sorted(dictionary.items())) 
        # print(sorted_terms)
        file_path = os.path.join(self.index_dir, f"block_{self.block_counter}.pkl")
        with open(file_path, "wb") as f:
            pickle.dump(sorted_terms, f)
        
        self.block_counter += 1

    def merge(self):
        files = sorted(os.listdir(self.index_dir))
        
        while len(files) > 1:
            new_files = []
            print(f"--- Iniciando nivel de fusión con {len(files)} bloques ---")
            
            # Fusionar los archivos en pares
            for i in range(0, len(files), 2):
                if i + 1 < len(files):  # Si hay un par para combinar
                    buffer1_path = os.path.join(self.index_dir, files[i])
                    buffer2_path = os.path.join(self.index_dir, files[i + 1])

                    with open(buffer1_path, "rb") as f:
                        dict1 = pickle.load(f)
                    with open(buffer2_path, "rb") as f:
                        dict2 = pickle.load(f)
                    
                    # Fusiona los dos bloques y mantiene el orden alfabético
                    merged_dict = self.merge_blocks(dict1, dict2)
                    
                    # Guardar el bloque fusionado en un nuevo archivo
                    new_file_path = os.path.join(self.index_dir, f"merged_{self.currentL}_{self.cuntermerged}.pkl")
                    with open(new_file_path, "wb") as f:
                        pickle.dump(merged_dict, f)
                    
                    new_files.append(f"merged_{self.currentL}_{self.cuntermerged}.pkl")
                    self.cuntermerged += 1
                else:
                    # Si no hay par para el último bloque, simplemente lo movemos al siguiente nivel
                    new_files.append(files[i])

            # Actualizar lista de archivos para el próximo nivel
            files = new_files
            self.currentL += 1
            print(f"--- Nivel {self.currentL} completado con {len(files)} bloques ---")

    def merge_blocks(self, dict1, dict2):
        merged_dict = {}
        iter1 = iter(sorted(dict1.items()))
        iter2 = iter(sorted(dict2.items()))

        term1, postings1 = next(iter1, (None, None))
        term2, postings2 = next(iter2, (None, None))

        while term1 is not None or term2 is not None:
            if term1 is not None and (term2 is None or term1 < term2):
                merged_dict[term1] = postings1
                term1, postings1 = next(iter1, (None, None))
            elif term2 is not None and (term1 is None or term2 < term1):
                merged_dict[term2] = postings2
                term2, postings2 = next(iter2, (None, None))
            else:
                combined_postings = postings1.copy()
                for doc_id, tf in postings2.items():
                    combined_postings[doc_id] = combined_postings.get(doc_id, 0) + tf
                merged_dict[term1] = combined_postings
                term1, postings1 = next(iter1, (None, None))
                term2, postings2 = next(iter2, (None, None))

            # Si excedemos el límite de memoria, escribimos en disco
            if sys.getsizeof(merged_dict) >= self.disk_limit:
                self.write_block_to_disk(merged_dict)
                merged_dict.clear()

        return merged_dict


             

s = SPIMI()  
s.spimi_invert(pairs_df)
s.merge()
```

```python
import os
import pickle

def print_full_index(index_dir):
    full_index = {}

    # Recorremos los archivos de los bloques fusionados en la carpeta
    files = sorted(os.listdir(index_dir))

    for file in files:
        if file.endswith(".pkl"):  # Asegurarnos de que es un archivo pickle
            block_path = os.path.join(index_dir, file)

            # Abrimos el archivo pickle y cargamos el bloque
            with open(block_path, "rb") as f:
                block = pickle.load(f)
                
                # Fusionamos los términos de este bloque con el índice completo
                for term, postings in block.items():
                    if term not in full_index:
                        full_index[term] = postings
                    else:
                        # Si el término ya existe, combinamos las frecuencias de los docID
                        for doc_id, freq in postings.items():
                            if doc_id in full_index[term]:
                                full_index[term][doc_id] += freq
                            else:
                                full_index[term][doc_id] = freq

    # Ahora 'full_index' contiene todos los términos y sus frecuencias
    # Puedes imprimirlo o retornar el diccionario completo
    for term, postings in full_index.items():
        print(f"Term: {term}")
        for doc_id, freq in postings.items():
            print(f"  DocID: {doc_id}, TF: {freq}")

    return full_index

full_index = print_full_index(s.index_dir)
```

```python
import os
import pickle

def print_full_index_and_check_order_twocriterios(index_dir):
    full_index = {}
    last_term = None  # Variable para verificar el orden de términos

    # Recorremos los archivos de los bloques fusionados en la carpeta
    files = sorted(os.listdir(index_dir))

    for file in files:
        if file.endswith(".pkl"):  # Asegurarnos de que es un archivo pickle
            block_path = os.path.join(index_dir, file)

            # Abrimos el archivo pickle y cargamos el bloque
            with open(block_path, "rb") as f:
                block = pickle.load(f)
                
                # Fusionamos los términos de este bloque con el índice completo
                for term, postings in block.items():
                    # Verificar si los términos están en orden alfabético
                    if last_term is not None and term < last_term:
                        print("El índice no está ordenado en términos.")
                        print(f"Error: {term} aparece después de {last_term}")
                        return

                    # Verificar si los `doc_id` están en orden ascendente
                    last_doc_id = None
                    for doc_id in sorted(postings):
                        if last_doc_id is not None and doc_id < last_doc_id:
                            print("El índice no está ordenado en `doc_id` dentro de un término.")
                            print(f"Error en el término {term}: {doc_id} aparece después de {last_doc_id}")
                            return
                        last_doc_id = doc_id

                    last_term = term  # Actualizar `last_term` para la siguiente comparación

                    # Combinar postings en el índice completo
                    if term not in full_index:
                        full_index[term] = postings
                    else:
                        for doc_id, freq in postings.items():
                            if doc_id in full_index[term]:
                                full_index[term][doc_id] += freq
                            else:
                                full_index[term][doc_id] = freq

    print("El índice está ordenado.")
    return full_index

# Ejecuta la función
full_index = print_full_index_and_check_order_twocriterios("index_blocks")
```

```python
import os
import pickle

def print_full_index_and_check_order(index_dir):
    last_term = None  # Variable para verificar el orden de términos

    # Recorremos los archivos de los bloques fusionados en la carpeta
    files = sorted(os.listdir(index_dir))

    for file in files:
        if file.endswith(".pkl"):  # Asegurarnos de que es un archivo pickle
            block_path = os.path.join(index_dir, file)

            # Abrimos el archivo pickle y cargamos el bloque
            with open(block_path, "rb") as f:
                block = pickle.load(f)
                
                # Verificar que los términos están en orden alfabético
                for term in block.keys():
                    if last_term is not None and term < last_term:
                        print("El índice no está ordenado en términos.")
                        print(f"Error: '{term}' aparece después de '{last_term}'")
                        return

                    last_term = term  # Actualizar `last_term` para la siguiente comparación

    print("El índice está ordenado en términos.")

# Ejecuta la función
print_full_index_and_check_order("index_blocks")
```
